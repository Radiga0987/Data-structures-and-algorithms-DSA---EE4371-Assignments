##### Show that mergesort algorithm has a run time of O(nlogn) for an n element sequence when n is not a power of 2    

###### To prove:    
T(n)=O(nlogn) where T(n) is the run time for the mergesort algorithm  
Hence we need to show:  
**T(n) $\leq$ Cnlogn**  for all n $\geq$ 1  and C is some arbitrary constant 
  
###### Proof:  
The run time T(n) is given by **T(n)=T($\left\lfloor{n/2}\right\rfloor$)+ T($\left\lceil{n/2}\right\rceil$) + bn**   
Here, $\left\lfloor{}\right\rfloor$ represents floor function and $\left\lceil{}\right\rceil$ represents ceil function  
The above given run time is derived by looking at how mergesort works.The array of n elements is split in between resulting   
in 2 new arrays in which one has length $\left\lfloor{n/2}\right\rfloor$ and other has length $\left\lceil{n/2}\right\rceil$ the merge part of mergesort takes "bn" time where  
b is an arbitrary constant.Therefore for sorting each of the 2 new arrays, it will take T($\left\lfloor{n/2}\right\rfloor$), T($\left\lceil{n/2}\right\rceil$) respectively and the  
sum will be the total time.

This question can be solved by the *method of induction*. 

**Base case** : When n=1 , T(1)=0 since we dont need to do any work and array is already sorted.      
$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$When n=1 , Cnlogn= Clog1 = 0.  
$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$Therefore, T(n) = Cnlogn when n=1  
$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$Hence base case is proved  

**Induction hypothesis** : Assume that T(k) $\leq$ Cklogk is true for all k < n 

**Inductive step** :  We need to prove T(k) $\leq$ Cklogk is true for k = n   
  We take the arbitrary constant C as **C=2b**  
  We have already shown that:  
  T(n) = T($\left\lfloor{n/2}\right\rfloor$)+ T($\left\lceil{n/2}\right\rceil$) + bn  
  Therefore,using above equation,  
  T(n) $\leq$ 2b$\left\lfloor{n/2}\right\rfloor$log$\left\lfloor{n/2}\right\rfloor$ + 2b$\left\lceil{n/2}\right\rceil$log$\left\lceil{n/2}\right\rceil$ + bn  
  T(n) $\leq$ 2b($\left\lfloor{n/2}\right\rfloor$+$\left\lceil{n/2}\right\rceil$)log$\left\lceil{n/2}\right\rceil$ + bn $\;\;\;$ (Since $\left\lfloor{n/2}\right\rfloor$ $\leq$ $\left\lceil{n/2}\right\rceil$)  
  using the formula- $\left\lfloor{n/2}\right\rfloor$+ $\left\lceil{n/2}\right\rceil$ = n  
  T(n) $\leq$ 2bnlog$\left\lceil{n/2}\right\rceil$ + bn   
  T(n) $\leq$ 2bnlog(2n/3) + bn  $\;\;\;\;$    (Since $\left\lceil{n/2}\right\rceil$ $\leq$ 2n/3 for all n $\geq$ 2)  
  T(n) $\leq$ 2bnlogn - bn(2log3/2 - 1)  
  The second term is negative and hence we can write,  
  T(n) $\leq$ 2bnlogn  
  (or)  
  T(n) $\leq$ Cnlogn
  
  Since we have proved T(k) $\leq$ Cklogk for k=n , by induction, we can say that  
  **T(n) $\leq$ Cnlogn for all n $\geq$ 1**

  Therefore **T(n) = O(nlogn)**  
  **Hence ,the run time of mergesort is O(nlogn) even when n is not a power of 2**

We can make sure that quicksort on a sequence runs in $\Omega(n^{2})$  by designing the sequence such that,it always  
reults in the worst case scenario.The worst case scenario for quick sort results in $\Omega(n^{2})$.This is the idea  
that is required to solve this question.
The worst case scenario of quick sort happens when the pivot is of maximum or minimum value.In this case,the partition    
will result into 2 subarrays one of length n-1 and other with length 0(no elements).  
So the recurrence relation is:  

T(n)=T(n-1) + bn $\;\;\;\;$(b is a constant)  
$\implies$ T(n)=T(n-2) + b(n-1) + bn  and so on  

We get,  
T(n)=b(1 + 2 + 3 + .... + (n-1) + n)  
$\implies$T(n)= $(bn(n-1))/2$  

$\therefore$ **T(n) = $\Omega(n^{2})$**  In the worst case scenario




#### Description of the sequence
To solve this question, our sequence should be of the following type:   
The first requirement is that the maximum/minimum element is in the middle of the list($\left\lfloor{n/2}\right\rfloor$ index).This is the basic requirement.

**The basic logic used here is that quick sort performs the worst when the pivot element is maximum or minimum

##### Describing and analyzing  an efficient method to remove duplicates from a collection A of n elements

##### Code explained
This question has been solved using mergesort and then some basic code to remove duplicates  
from a sorted list.
I have written two functions mergesort and merge,where mergesort uses the merge function in     
its code to sort a given list in ascending order.So, after applying mergesort on the user given   
list, we have the list in sorted order but now we have lost the order in which the original list    
had its elements in.Therefore,to prevent this, under the duplicateremover function, i have made a    
2 Dimensional list containing 2 coloumns and number of rows equal to number of elements in the list.    
The first coloumn has the elements of the given list and the second coloumn has the indices of the    
corresponding elements.Now, we can sort this 2-D list with respect to the first coloumn.When we sort     
this 2-D list using the mergesort function, we have the elements of the list in sorted order along   
with their original index position.  So we can use this index positions in the second coloumn later on   
to preserve the order of the original list.Now that we have a the sorted list, we use 2 while statements    
and append the first element into a new list and then we check whether the element we are currently at  
and the next element are equal.If they are equal, then we increment the i value used so that we skip appending    
the same element again.We use a while statement to do so.  
Therefore, at the end of this process, we have a list with no duplicates.Now we mergesort our new list    
with respect to the indices we had stored before, so that the elements regain their original relative order.  

###### Note: The mergesort function defined here always sorts with respect to the first coloumn of the 2-D list.
Therefore, at the ending we have swapped the 2 coloumns so that the merge sort sorts with respect to the   
indices.Then, the elements in the second coloumn of this list is printed.

##### Time Complexity  
The mergesort function has order of $O(nlogn)$.  
This is shown by solving the recurrence relation:  
T(n)=2T(n/2) + n    
$\implies$T(n)=4T(n/4)+ 2n and so on  

$\therefore$T(n)=$2^{i}$T(n/$2^{i}$) + in

taking n/$2^{i}$=1,we get  
i=logn  

therefore,    
T(n)=$2^{logn}$T(1)+nlogn $\;\;\;$   ($2^{logn}$=n)  

So mergesort has time complexity of the order $O(nlogn)$  


The while statement has a time complexity of the order $O(n)$.    
Therefore, since $O(nlogn)$ > $O(n)$,  
time complexity of the code is $O(nlogn)$ 

```python
def merge(L,R,k):      #Merge function used by mergesort function to merge the lists L,R
    r=[]               #r is the list which contains the merged lists at the end of the function
    i=0
    j=0
    for k in range(0,k):    #For loop with k iterations to append k total elements to the list r from the lists L,R 
        if i<len(L) and j<len(R): #If the current values of i, j point to some elements in L,R ,we compare them and then append
            if L[i]<R[j]:           
                r.append(L[i])
                i+=1
            else:
                r.append(R[j])
                j+=1
        elif j<len(R):  #When all the elements in list L have already been appended, we append the elements remaining in R.
            r.append(R[j])
            j+=1
        else:           #If all the elements in list R have already been appended, we append the elements remaining in L. 
            r.append(L[i])
            i+=1
    return r           #Merged list is returned.
    
    
def mergesort(li):       #Mergesort function for sorting in O(nlogn)
    
    length=len(li)
    if length<=1:        #If the length of list is 1, it means we have divided till base case and hence, the list is returned.
        return(li)
    L=mergesort(li[0:length//2])       #Mergesort is recursively called in the following 2 cases until base case is reached.
    R=mergesort(li[length//2:length])
    return(merge(L,R,length))          #L, R are merged recursively until it gives the full sorted list which is the output.



def duplicateRemover(li):
    lst=[]
    for i in range(0,len(li)):  
        lst.append([li[i],i])   
    #lst is a 2-D list containing [element, index in original list] as each of its coloumns for a particular row 
    l1=mergesort(lst)  #l1 is lst but with mergesort applied to first coloumn of lst.
    l2=[]              #l2 is the list which will contain elements without the duplicates.
    i=0
    while(i<len(l1)):       
        #Using a while loop, the elements in l1 which are sorted are traversed over and the duplicates are not appended to l2.
        l2.append(l1[i])  #Elements are appended to l2.
        if i<len(l1)-1:
            while(l1[i][0]==l1[i+1][0]):  
        #If the current element and the next element are equal, i is incremented so that it skips appending the next element
                if(i<len(l1)-2):
                    i+=1
                else:       #If end of list is reached, we break out of the loop.
                    i+=1
                    break
        i+=1  #General increment of i independent of whether the current element and next one are equal
        
#This removal of duplicates takes O(n) but the sorting takes O(nlogn).Therefore time complexity of the code is O(nlogn)
    
    for j in l2:       #The coloumns of l2 are swapped so that we can now mergesort with respect to the indices.
                       #This is done to preserve the original order of the list.
        j[0],j[1]=j[1],j[0]

    l3=mergesort(l2)   #mergesort with respect to indices.
    l4=[]
    for i in range(0,len(l3)):
        l4.append(l3[i][1])  #The second coloumn which contains the elements now is appended to l4.
    print(l4)                #l4 contains elements without duplicates and order of list is preserved.This is printed.
    
    


#Few Test Cases:
print("Case 1:")   #List with few duplicates
duplicateRemover([1,1,10,2,193,2,-19,28,-19]) 

print("\n")

print("Case 2:")
duplicateRemover([5,5,5,5,1,10,-20,1,1,1])

print("\n")

print("Case 3:") #List with no duplicates
duplicateRemover([10,100,293])

                 
```

##### Describe a method for sorting an array A of n integers in the range [0,$n^{2}$-1] in O(n) time

This question has been solved using a sorting method known as Radix sort.Radix sort uses another sorting method  
known as counting sort as a sub function. I have implemented Radix sort with base n, where n is the length of the  
input list.When each number in the list is converted to base n, they become 2 digit numbers(in base n) , since all  
the numbers are in the range [0,$n^2-1$] .  
To sort the given list, first we apply counting sort with respect to the least significant digit (right digit) for the  
numbers in base n (2 digit numbers) and after this, we apply counting sort with respect to the most significant digit   
(left digit) for the numbers in base n (2 digit numbers) .

#### Run time of counting sort:  (In this code only, this is not a general proof for run time of counting sort)
1)2 lists outputlst and countlst of length n have been made in which all elements are 0.  
$\;\;$This takes O(n) time.  

2)After this , there are 4 for loops each taking O(n) time.  

Therefore, 
Run time= O(n) + O(n) + O(n) + O(n) + O(n)

**Run time = O(n)** for counting sort function

#### Run time of the entire  code for sorting  :
We apply counting sort twice, first with Placedecider=1 and then Placedecider= n.  
This is done so that the first sort happens with respect to the right digit and then second sort happens      
with respect to the second digit of the numbers in base n form.

Since counting sort is used only twice here to give the final sorted list, the run time will be given as:    
Run time = O(n) + O(n)  
$\therefore$**Run time = O(n)** for the entire sorting code.

**Important note**:   
**This method of sorting is stable.That is, radix sort is a stable sort.This means that the relative orders  
of elements with same digits(either right or left)are preserved while sorting with respect to  a specific digit.**  
This is required as we first sort with respect to the right digit.When we do this, if 2 numbers have the same right digit,    
then we do not want their relative orders to change while sorting.

#### Explanation of how countingsort works :

1)Consider a input list lst of n elements with numbers in the range [0,k], where n and k are positive.   
$\;\;$This list can contain duplicates too.  

2)The count/frequency of each distinct element is stored in another list and we name this list as count  
$\;\;$which has size as k+1.Let x be an element in the input list (lst) such that its frequency is stored  
$\;\;$at count[x].   

3)Now, we update the count array so that element at each index of count, say i , is given by :      
$\;\;\;\;\;\;\;$ count[i]=$\sum$ count[u] $\;\;\;\;$where 0$\leq$ u $\leq$ i  

4)The updated count list gives the index of each element of list lst in the sorted sequnece.Let the sorted  
$\;\;$ sequence be stored in an output list(outputlst) of size n which is initialised to zeroes.

5)Now we add each element from input list lst to the output list as follows:  
$\;\;$Set i=0  
$\;\;$a)t=lst[i]  
$\;\;$b)add t to outputlst[count[t]-1]  
$\;\;$c)decrease count[t] by 1   
$\;\;$d)increase i by 1 and then jump to step a 
$\;\;$We do this till i=n-1

Now,outputlst will contain the sorted list

**The code for this question is given below**

```python

def countingsort(lst,PlaceDecider):  #This is the counting sort function for sorting with respect to a particular digit.
    n=len(lst)          #Length of lst is stored in n
    outputlst=[0]*n     #Initialising 2 lists of length n with all elements as 0
    countlst=[0]*n
    
    for i in range(0,n): 
        placevalue=(lst[i]//PlaceDecider)%n   
        #placevalue will store the right digit if PlaceDecider=1 and left digit if PlaceDecider=n=len(lst)
        countlst[placevalue]+=1 
        #The value stored in index=placevalue is incremented
        
    for i in range(1,n):  
        countlst[i]+=countlst[i-1]  
        #Count is changed to a cumulative sum in order to know which index will have the elements in the output list.
        
    for i in range(n-1,-1,-1):  #We go from n-1 to 0 here in order to make the sort stable.
                                #Relative orders of elements preserved if they have same digits(either right or left decided by PlaceDecider)
        placevalue=(lst[i]//PlaceDecider)%n 
        #placevalue will store the right digit if PlaceDecider=1 and left digit if PlaceDecider=n=len(lst)
        #This is same as before.
        
        outputlst[countlst[placevalue]-1]=lst[i]    
        #The unsorted list(lst) is used to store its sorted form in outputlst with respect to right or left digit
        #Sorting with respect to left or right digit depends on PlaceDecider as stated before.
        #The index of outputlst in which lst[i] is stored is decided by the cumulative values stored in countlst. 
        
        countlst[placevalue]-=1
        
    #At the end of the above for loop, outputlst will have the sorted list with respect to right or left digit.
    for i in range(0,n):
        lst[i]=outputlst[i]   #The sorted list in outputlst is stored into lst to get the sorted array.
                              #Again, here sorted is with respect to either right or left digit(Decided by PlaceDecider)
        
def Mainsortfunction(lst):      #This is the mainsorting function,Basically radix sort.
    
    countingsort(lst,1)         #First,we use counting sort to sort with respect to right digit(LSD) in base n  
    countingsort(lst,len(lst))  #Then, we use counting sort to sort with respect to left digit(MSD) in base n
                                #After above 2 steps, the input list lst will be sorted in O(n) time
    
    print("The sorted list is:")
    print(lst,"\n")            #Sorted list is printed
    

#All the elements of the list should be in the range [0,(n^2)-1] for this sort to work.Here, n is length of list.    
    
#Few Test Cases:   
#Case 1
print("Case 1:")
l=[1,4,45,6,3,87,36,19,3,91,29]
Mainsortfunction(l)

#Case 2
print("Case 2:")
l=[10,2,9,29,33,29]
Mainsortfunction(l)

#Case 3
print("Case 3:")
l=[100,28,57,92,43,10,43,1,93,111,43,134]
Mainsortfunction(l)
```

##### Show that  quicksorts best case running time is $\Omega(nlogn)$  

Let us take take T(n) as the run time for quicksort on an n element array.  
Let q be the index of pivot after placing it in the correct place.  

We can write T(n) as:  
**T(n)= T(q)+T(n-q-1) + O(n)**   
The array is partitioned at index q and therefore we have to recursively call quicksort to perform  
the same steps for the 2 parts with lengths q and n-q-1 and hence we get the above equation.  

We know that in the best case, the array will be partitioned into 2 parts of equal length (n/2).  
#### Important note:  
**This is assuming n is a power of 2 . If n is not a power of 2 , we have to take ceil and floor  
to prove this, but we will get same answer.For simplicity of proof, we can take n is power of 2**  
**If we dont take n as a power of 2 , then the proof is almost the same as the proof in question 1 of this assignment.** 

The fact that this is best case can be proved by some simple differentiation of T(q)+T(n-q-1)   
with respect to q.We take T(q) as qlogq and T(n-q-1) as (n-q-1)log(n-q-1) and then differnetiate  
and then equate to 0.We get q=n/2 and using second derivative test, we get that this will give the  
minimum value and hence the best case run time. 

In this case,   
**T(n)=2T(n/2) + O(n)**  
$\implies$ T(n)=4T(n/4) + O(n)  $\;\;\;\;$(Since using above equation we can write T(n/2)=2T(n/4)+O(n))  
$\;\;\;\;\;\;\;\;$and so on  
In general,  
**T(n)=$2^{i}$T(n/$2^{i}$) +O(n)**  

Now taking i=$log_2 n$  and T(1)=1 we get,  
**T(n)=nlogn +O(n)**  
This is the run time for the best case of quick sort.  

**Therefore,  
$\Omega$(nlogn) is the best case run time of quick sort**
