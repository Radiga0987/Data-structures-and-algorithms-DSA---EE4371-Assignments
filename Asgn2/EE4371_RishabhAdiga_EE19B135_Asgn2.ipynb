##### Algorithm for finding 10 largest elements in a sequence of size n:

##### Code explained
The Question demands us to find the ten largest elements in a given sequence.  
The function tenlargest(lst) has an initial if else statement for taking care of edge cases.(Number of elements less than 10)  
  
Independent of the size of the given sequence,the given code uses a sorting technique in order to  
get the 10 largest elements of the list.But,instead of sorting the entire list of elements, we sort   
such that we find the 10 largest elements first and append them into a different list.We do not need   
to sort the remaining elements as we only need the 10 largest elements.This reduces the the time complexity  
from $n^{2}$ to $n$.

We take 10 iterations(as we need 10 largest elements), and in each iteration we find the largest element   
and append it to the list lstTenlargest[$\;$].After finding the largest element in a particular iteration,  
we set the corresponding index of a different list of zeros(check[$\;$]) to 1.This is done so that the code doesnt      
check the same element for finding the next maximum value.So only if check[index]==0,that element should be     
considered for finding maximum in that particular iteration.  

This is done 10 times to get 10 maximum values of the list which are added to the list lstTenlargest[].  

##### Time complexity 
The code has 2 nested for loops and lot of assignments of variables and a lot of if, else statements.    
The 2 nested for loops contains an outer for loop which runs for ten iterations and an inner for loop   
which runs for $n$ iterations.Therefore, the time complexity of the for loops will be of the form   
$10n$.$\implies$$O(n)$.If and else statements are of order $O(1)$.

Therefore,the run time,T(n)= $an+b$ where a, b are constants which can be found but requires extensive analysis.  
Therefore, Time complexity for this code is of the order $O(n)$


```python

def tenlargest(lst):    #Function for finding the ten largest elements of a list
    if len(lst)==0: 
        print("There are no elements in the given list")
        return
    if len(lst)<10:     #This is for edge cases.That is when there is less than 10 elements                  
        print("There are only",len(lst),"elements in the given list.Therefore,only",len(lst),"largest elements exist.")
        print("The largest elements are:")
    else:
        print("The 10 largest elements are :")
    
    lstTenLargest=[]    #List to store the ten largest elements
    check=len(lst)*[0]  #List to make sure that we dont add the same largest element multiple times.Initially all elements are 0.
    
    for i in range(0,10):  
   
        #For loop with 10 iterations because we need only the ten largest elements.
        #Hence we do not need to make the for loop for n iterations.
       
        maximum= None   #maximum initialised to none(no value).
        c=0         #c variable is used to change check[c] to 1 where c is the index corresponding to the largest element picked
        for j in range(0,len(lst)):
            #For loop to go over all elements to find the maximum.
            if maximum is not None:  
                if lst[j]>maximum:      #This code finds the maximum in the list(not including previous maximums.)
                    if check[j]==0:
                        maximum=lst[j]  #When a new maximum value is found, c takes value equal to the index of that element.
                        c=j             
            else:
                if check[j]==0:       #For the first iteration of the inner for loop, the else statement is written
                    maximum=lst[j]    #That is when maximum=None.
                    c=j
        
        check[c]=1     #For each iteration of the outer forloop,check[c] is made 1 so that we dont append the same maximum again 
        lstTenLargest.append(maximum) #The maximum value is appended to the output list
    
    #Order of this code will be O(n) as the outer for loop has 10 iterations and inner has n iterations(Therfore that gives 10n)
    
    for i in lstTenLargest:   #Printing the 10 largest elements.   
        if i!= None:
            print(i)


            
#Few Test Cases:
print("Case 1:")
tenlargest([5,-20,100,29,30,16,-2,4,10,56,1092,-902,192]) 

print("\n")

print("Case 2:")
tenlargest([-20,-21,-12,-30,-1,-9,10,20,10023,-3,2,465,1,10])

print("\n")

print("Case 3:")
tenlargest([1,10,23,-1,-2])

                 
```

##### Multiplying the binary numbers 10011011 and 10111010 using divide and conquer:

##### Code explained
This question has been solved by implementing divide and conquer along with kartsubas    
method of multipying 2 numbers which results in $O(n^{1.59})$.    
For simplicity and understandability of code, the input numbers which are binary      
numbers are converted to decimals using the function $BinToDec$ and then the multiplying     
algorithm has been coded for the decimal numbers.   
If 2 numbers X and Y are to be multiplied,then split X and Y each into 2 parts.   
X=Xl*($10^{n/2}$)+Xr    
Y=Yl*($10^{n/2}$)+Yr    
Where n is the number of digits of the largest number in X,Y.    
Then X*Y=$(Xl*Yl)$*$10^{n}$ + $((Xl+Xr)*(Yl+Yr)-(Xl*Yl)-(Xr*Yr))$*$10^{n/2}$ +$Xr*Yr$  
Using Karatsubas technique for multiplication of 2 numbers and divide and conquer,  
we can recursively call the same function until we reach base case and then multiply    
the higher cases using karatsubas technique.  

##### Time complexity
Conversion of numbers from binary to decimal and decimal to binary is of the order of $O(n)$.  
This is because we visit each digit in the number once and then do the operation for conversion.

Karatsubas method of multiplication is of the order of $O(n^{1.59})$ or to be precise, $O(n^{log_2(3)})$.   
We get the reccurence relation as :  
T(n)=3T(n/2)+O(n)    
$\implies$T(n)=9T(n/4)+O(n) and so on.  

$\therefore$T(n)=$3^{i}$T(n/$2^{i}$)+O(n)  

taking i=$log_2 n$, we get  
n/$2^{i}$=1  

$\therefore$T(n)=$3^{log_2 n}$T(1)+O(n)  
T(1)=1

Hence,  
T(n)=$3^{log_2 n}$+O(n)  
T(n)=$n^{log_2 3}$+O(n) 


$\therefore$Time complexity for this code is of the order $O$($n^{log_2 3}$) 





```python
def BinToDec(a): #Function to convert binary number(base2) to decimal number(base10)
    l=list(str(a))
    dec=0
    for i in range(0,len(l)):
        dec+=int(l[i])*(2**(len(l)-i-1))
        #Summing up the digits multiplied by 2power(quantity for conversion corresponding to the index)
    return(dec) #Returning the decimal number



def DecToBin(a):   #Function for converting decimal number(base10) to binary number(base2) 
    if a==0:  #Edge case.
        return(a)
    binlist=[]
    while True:  #Making an infinite loop which will execute as long as the else statement is not accessed.
        if a>=1:
            binlist.insert(0,a%2) #Prepending a%2 to binlist.
            a=a//2  #Integer division of a by 2.
        else:   #When  a<1, we break out of the while loop.
            break
    #The above code is the mathematical equivalent of converting decimal number to binary number.
    s = [str(i) for i in binlist] 
    binaryno = int("".join(s)) #Converting string s into an integer number containing 1s and 0s which is the binary form. 
    return(binaryno) 
    


def div_conquer_karatsuba(p,q): #Function for multiplying to decimal numbers using divide and conquer and karatsubas technique.
    if len(str(p))==1 or len(str(q))==1: #Test for base case, when either of p, q has only one digit.
        return(p*q)

    n = max(len(str(p)),len(str(q)))  #maximum of number of digits of the 2 numbers p,q.
    
    plist=list(str(p))
    qlist=list(str(q))

    w= p//(10**(n//2))  # w will store the left half of p(similar to Xl in the above given explaination of code)
    x= p%(10**(n//2))   # x will store the right half of p(similar to Xr in the above given explaination of code)
    y= q//(10**(n//2))  # y will store the left half of q(similar to Yl in the above given explaination of code)
    z= q%(10**(n//2))   # z will store the right half of q(similar to Yr in the above given explaination of code)

    term1= div_conquer_karatsuba(x,z)  #Same as finding Xr*Yr
    term2= div_conquer_karatsuba((w+x),(y+z))  #Same as finding (Xl+Xr)*(Yl+Yr)
    term3= div_conquer_karatsuba(w,y) #Same as finding Xl*Yl

    return ((10**(2*(n//2)))*term3)+((10**(n//2))*(term2-term3-term1))+term1 #Applying karatsubas method of multiplication.


def binmultiply(a,b): 
    #Main function which uses all the previous functions and finds the product of binary numbers a and b.
    c=BinToDec(a) #converts a to decimal and stores in c
    d=BinToDec(b) #converts b to decimal and stores in d
    ans=div_conquer_karatsuba(c,d) # c*d in decimal system is stored in ans
    return(DecToBin(ans))  #ans is converted to binary and returned
    

#For given question, we need to multiply the binary numbers 10011011 and 10111010
print(binmultiply(10011011,10111010))
```

##### Removing duplicates from an array in  O(nlogn) time:

##### Code explained
This question has been solved using mergesort and then some basic code to remove duplicates  
from a sorted list.
I have written two functions mergesort and merge,where mergesort uses the merge function in     
its code to sort a given list in ascending order.So, after applying mergesort on the user given   
list, we have the list in sorted order but now we have lost the order in which the original list    
had its elements in.Therefore,to prevent this, under the duplicateremover function, i have made a    
2 Dimensional list containing 2 coloumns and number of rows equal to number of elements in the list.    
The first coloumn has the elements of the given list and the second coloumn has the indices of the    
corresponding elements.Now, we can sort this 2-D list with respect to the first coloumn.When we sort     
this 2-D list using the mergesort function, we have the elements of the list in sorted order along   
with their original index position.  So we can use this index positions in the second coloumn later on   
to preserve the order of the original list.Now that we have a the sorted list, we use 2 while statements    
and append the first element into a new list and then we check whether the element we are currently at  
and the next element are equal.If they are equal, then we increment the i value used so that we skip appending    
the same element again.We use a while statement to do so.  
Therefore, at the end of this process, we have a list with no duplicates.Now we mergesort our new list    
with respect to the indices we had stored before, so that the elements regain their original relative order.  

###### Note: The mergesort function defined here always sorts with respect to the first coloumn of the 2-D list.
Therefore, at the ending we have swapped the 2 coloumns so that the merge sort sorts with respect to the   
indices.Then, the elements in the second coloumn of this list is printed.

##### Time Complexity  
The mergesort function has order of $O(nlogn)$.  
This is shown by solving the recurrence relation:  
T(n)=2T(n/2) + n    
$\implies$T(n)=4T(n/4)+ 2n and so on  

$\therefore$T(n)=$2^{i}$T(n/$2^{i}$) + in

taking n/$2^{i}$=1,we get  
i=logn  

therefore,    
T(n)=$2^{logn}$T(1)+nlogn $\;\;\;$   ($2^{logn}$=n)  

So mergesort has time complexity of the order $O(nlogn)$  


The while statement has a time complexity of the order $O(n)$.    
Therefore, since $O(nlogn)$ > $O(n)$,  
time complexity of the code is $O(nlogn)$  

```python

def merge(L,R,k):      #Merge function used by mergesort function to merge the lists L,R
    r=[]               #r is the list which contains the merged lists at the end of the function
    i=0
    j=0
    for k in range(0,k):    #For loop with k iterations to append k total elements to the list r from the lists L,R 
        if i<len(L) and j<len(R): #If the current values of i, j point to some elements in L,R ,we compare them and then append
            if L[i]<R[j]:           
                r.append(L[i])
                i+=1
            else:
                r.append(R[j])
                j+=1
        elif j<len(R):  #When all the elements in list L have already been appended, we append the elements remaining in R.
            r.append(R[j])
            j+=1
        else:           #If all the elements in list R have already been appended, we append the elements remaining in L. 
            r.append(L[i])
            i+=1
    return r           #Merged list is returned.
    
    
def mergesort(li):       #Mergesort function for sorting in O(nlogn)
    
    length=len(li)
    if length<=1:        #If the length of list is 1, it means we have divided till base case and hence, the list is returned.
        return(li)
    L=mergesort(li[0:length//2])       #Mergesort is recursively called in the following 2 cases until base case is reached.
    R=mergesort(li[length//2:length])
    return(merge(L,R,length))          #L, R are merged recursively until it gives the full sorted list which is the output.



def duplicateRemover(li):
    lst=[]
    for i in range(0,len(li)):  
        lst.append([li[i],i])   
    #lst is a 2-D list containing [element, index in original list] as each of its coloumns for a particular row 
    l1=mergesort(lst)  #l1 is lst but with mergesort applied to first coloumn of lst.
    l2=[]              #l2 is the list which will contain elements without the duplicates.
    i=0
    while(i<len(l1)):       
        #Using a while loop, the elements in l1 which are sorted are traversed over and the duplicates are not appended to l2.
        l2.append(l1[i])  #Elements are appended to l2.
        if i<len(l1)-1:
            while(l1[i][0]==l1[i+1][0]):  
        #If the current element and the next element are equal, i is incremented so that it skips appending the next element
                if(i<len(l1)-2):
                    i+=1
                else:       #If end of list is reached, we break out of the loop.
                    i+=1
                    break
        i+=1  #General increment of i independent of whether the current element and next one are equal
        
#This removal of duplicates takes O(n) but the sorting takes O(nlogn).Therefore time complexity of the code is O(nlogn)
    
    for j in l2:       #The coloumns of l2 are swapped so that we can now mergesort with respect to the indices.
                       #This is done to preserve the original order of the list.
        j[0],j[1]=j[1],j[0]

    l3=mergesort(l2)   #mergesort with respect to indices.
    l4=[]
    for i in range(0,len(l3)):
        l4.append(l3[i][1])  #The second coloumn which contains the elements now is appended to l4.
    print(l4)                #l4 contains elements without duplicates and order of list is preserved.This is printed.
    
    


    
#Few Test Cases:
print("Case 1:")   #List with few duplicates
duplicateRemover([1,1,10,2,193,2,-19,28,-19]) 

print("\n")

print("Case 2:")
duplicateRemover([5,5,5,5,1,10,-20,1,1,1])

print("\n")

print("Case 3:") #List with no duplicates
duplicateRemover([10,100,293])

                 
```
