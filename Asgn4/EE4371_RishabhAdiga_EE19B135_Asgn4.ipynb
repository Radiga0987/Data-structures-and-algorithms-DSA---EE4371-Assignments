##### Show how to implement a stack using two queues. Analyze run times of the stack operations.

Stack is a data type that serves as a collection of elements, with two main principal operations:  
1)**Push**, which adds an element to the collection, and  
2)**Pop**, which removes the most recently added element that was not yet removed.  
The order in which elements come off a stack gives rise to its alternative name, **LIFO (last in, first out)**  

In queues , elements are added to the rear end of the queue and this operation is called **enqueue**(*in python this is done using **q.put(i)** , i is added)*.    
But here, elements are removed from the front end and this operation is called **dequeue**(*In python this is done using **q.get()** , element at front end is removed)*.    
Therefore, it is also called **FIFO (First in , first out)**.


For this question we need to use 2 queues to develop a stack.  
There are 2 methods we can use to implement this:   
**1)We can make the pop operation costly , or    
2)We can make the push operation costly**

#### Implementation of the stack:
Let us solve this question using method 1 **(pop operation less efficient)**:   
Let the 2 queues be q1 and q2.    
We shall use q1 to maintain ordering of the elements during operations and q2 is used for storing elements while popping.  

When push is used, the element being pushed is directly inserted into q1.    
When pop is used , elements in q1 are removed and inserted into q2 one by one until there is only one element or less.  
Then, this element that is left over is popped.(Basically the last element is the element to be popped , so to access it,    
the above method is used) . Also , when elements are removed from q1 and inserted into q2 , the ordering is not lost as  
both are FIFO.Moving on, now the entire queue elements are stored in q2 and q1 is empty.So now, we rename q1 as q2 and q2   
q1 so that the queue operations can be used again.The changing of names will take O(1) time as we are just swapping two   
pointers.

### Time complexity  

#### Push:  
Push operation in this method takes **O(1)** time as it is just one insert command into q1. 

#### Pop:  
Pop operation in this method takes **O(n)** time as it by doing n deletes and n-1 inserts and then the pointer swap takes  
O(1) time.**( O(n) + O(n-1) + O(1) = O(n) )**

### Pseudo code  

**Class Stack:   
$\;\;\;\;\;\;\;$Data:**  
$\;\;\;\;\;\;\;\;\;\;$Queue q1  
$\;\;\;\;\;\;\;\;\;\;$Queue q2    

$\;\;\;\;\;\;\;$**Function:**    
$\;\;\;\;\;\;\;\;\;\;$function Push()  
$\;\;\;\;\;\;\;\;\;\;$function Pop()

**function Push(i):**    
$\;\;\;\;\;\;\;\;$q1.put(i) 


**function Pop():   
$\;\;\;\;\;\;\;\;$while q1 has more than 1 element:**     
$\;\;\;\;\;\;\;\;\;\;\;\;\;$temporary=q1.get()  
$\;\;\;\;\;\;\;\;\;\;\;\;\;$q2.put(temporary)

$\;\;\;\;\;\;\;\;$**if q1 is not empty:**    
$\;\;\;\;\;\;\;\;\;\;\;\;\;$return q1.get()  
$\;\;\;\;\;\;\;\;\;\;\;\;\;$swap(q1,q2)  
$\;\;\;\;\;\;\;\;$**else:**  
$\;\;\;\;\;\;\;\;\;\;\;\;\;$"Stack is empty"

________
**Code for this has been implemented below using the Queue library/module**


```python
from queue import Queue

class Stack:                #Using a class for creating the stack.
    def __init__(self):
        self.q1=Queue()     #Creating 2 queues called q1 and q2
        self.q2=Queue()
        
    def pop(self):          #Defining the pop function for the stack using the two queues
        
        while self.q1.qsize() > 1:    #The pop function has been created using the method described earlier.
            a=self.q1.get()           #If number of elements in q1 are more than 1,then we move all to q2(except last element)
            self.q2.put(a)
            
        if self.q1.qsize()==1:        #After above step, only one element is left in q1(unless q1 was empty initially)
            popvalue=self.q1.get()    #This element is the popped element
            self.q1,self.q2 = self.q2, self.q1    #Names of q1 and q2 are swapped
            
        else:                       #If q1 was empty, then the following is executed
            popvalue=None
            print("Stack is empty")
            return popvalue
        
        print("Popped element is:", popvalue)
        return popvalue
    
    def push(self,pushvalue):   #Defining the push function for the stack using q1.
        self.q1.put(pushvalue)
        
    def display(self):          #Defining a function to display current status of the stack
        print("Stack is",list(self.q1.queue))
        

#Testing the code
        
stack1=Stack()     #Creating a Stack named stack1

stack1.push(2)     #2 and 9 are pushed into stack1
stack1.push(9) 

stack1.display()   #Current status of stack1 is printed

stack1.push(20)    #20 is pushed into stack1

stack1.display()   #Current status of stack1 is printed

stack1.pop()       #Topmost element of stack1 is popped.That is 20 is popped

stack1.display()   #Current status of stack1 is printed

stack1.pop()       #Topmost element of stack1 is popped.That is 9 is popped
stack1.pop()       #Topmost element of stack1 is popped.That is 2 is popped
stack1.pop()       #stack1 is empty and this is displayed

stack1.display()
```

Now that we know that y is an ancestor of x, we can say that x has to be in

##### Describe a non-recursive algorithm for enumerating all permutations of the numbers {1, 2, ...,n} using an explicit stack.

We are required to find out all possible permutations of the number {1,2,...,n} using  
an explicit stack.If we can find all permutations of numbers from 1 to k-1, we can  
easily find all permutations of numbers from 1 to k by just adding k in all possible  
locations of the permutations of 1 to k-1 one by one.This can be done by using .insert()  
command.  

#### Method to generate all permuations of numbers from 1 to n (Algorithm)
ðŸ ŠWe create an explicit stack named stack using a list.The end of the list is where elements are   
$\;\;\;$pushed and popped from so that it behaves like a stack.  

ðŸ ŠNow we push the number 1 into stack (**Using append command which acts like push for a stack**)  

ðŸ ŠNow we use a while loop with condition that stack is not empty.We pop the stack and get the  
$\;\;\;$top most permutation currently present in the stack.  

ðŸ ŠIf the length of the this permutation is **equal to n** then it means that this is one possible  
$\;\;\;$permuatation and hence we print it and end the if statement.  

ðŸ ŠElse, if the length of the this permutation is **not equal to n** , we have to insert the number  
$\;\;\;$ (length of top permutation)+1 to the top permutation in all possible locations of the list  
$\;\;\;$ one by one and push them back into the stack.

**This process is repeated until the while condition is not satisfied after which, all possible  
permutations of the numbers from 1 to n will be printed.**

**Note-The number of permutations is given by n! (n factorial)** 

The code for this is given below:  

```python
from copy import copy

def allpermutations(n):
    stack=[]               #Creating the stack
    stack.append([1])      #Pushing 1 into the stack (append behaves like push)
    
    while len(stack)!=0:
        topPerm=stack.pop()       #Popping the top most permutations and storing it in topPerm variable
        if len(topPerm)==n:       # If it has length=n then we have to print it.
            print(*topPerm,"\n")
            
        else:           
        #Else,we have to insert length of topPerm + 1 into all possible positions one by one and push these into the stack
        #The following for loop ensures this.
            for i in range(0,len(topPerm)+1):   
                temp=copy(topPerm)      #temp is just a copy of topPerm so that we dont alter topPerm as we use it again
                temp.insert(i,len(topPerm)+1)
                stack.append(temp)
                
                
#Test Case-all permutations of {1,2,3,4}
allpermutations(4)
#For n=4 , we will get 4!=24 permutations in the output.
```

##### Show that any n-node binary tree can be converted to any other n-node binary tree using O(n) rotations.  

Before solving this, we need to point out two important features of   
a binary tree.These are:  

**1)** There are two types of rotations that we can perform.These are left   
and right rotations.**Left rotation is an inverse operation of right rotation and     
vice versa.**  

**2) Every Binary tree with n nodes can be transformed into a right-going  
chain by using at most n-1 right rotations.**   
**Proof:**  
If we are given a binary tree which is already a right-going chain, then  
there is no need to do anything.Otherwise, we will have atleast one left-going   
edge in the tree.On this edge we can perform right rotation.After the rotation,  
the child node will move to the right of its parent node, which implies that  
the number of nodes in the right subtree is larger by 2.Also,for any node x which   
is a left child ,one right rotation on the parent of x will add that node to the   
chain and not remove any elements from the chain .Thus,the number of nodes  
in right subtree grows everytime we do the rotation. Hence , after at most n-1   
right rotations, the whole tree gets transformed into right-going  chain.  


### Solution for given question:  
Now, using the above 2 features, we shall prove that any n-node binary tree  
can be converted to any other n-node binary tree using O(n) rotations.  
Let us take two different binary search trees which are T1 and T2 with n nodes  
each. We need to prove that we can transform T1 into T2 with at most O(n)   
rotations.First, we transform T1 into a right going chain, which takes **at  
most n-1 rotations** as proved before.Now, we can transform T2 to the same  
right-going tree using **at most n-1 rotations** again. But since left rotations  
are inverses to right rotations, we can use left rotations on the the right-going  
tree to obtain tree T2.This obviously will take **at most n-1 rotations**.Therefore,  
we can convert any tree T1 to T2 and it will take **at most (n-1)+(n-1)= 2n-2  
rotations** to do this.**O(2n-2)=O(n)** and hence the transformation from tree T1  
to tree T2 can be done in O(n) rotations.
