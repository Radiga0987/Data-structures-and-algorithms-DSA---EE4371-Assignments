```python
#Question 1
#a
def squares_sum(n):    
    sum=0;
    for i in range(1,n):    #for loop which squares and adds the number in the current iteration to sum till nth number(n not included)
        sum+=i**2;
    print(sum);
#b
def squares_sum_odd(n):
    sum=0;
    for i in range(1,n,2): #Same as before but only odd numbers are squared and added(increment of 2 has been made after each iteration)
        sum+=i**2;
    print(sum);
```

```python
#Question 3
print("Enter the sequence of integers:")
l=list(map(int,input().split())) #Enter the sequence of integers in the console.


def Checkoddproduct(lst):
    for i in range(len(lst)):
        for j in range(i+1,len(lst)):
            if lst[i]!=lst[j]:   #If the for loop encounters 2 numbers in the sequence which are not equal and their product is 0,it returns true.
                if (lst[i]*lst[j])%2!=0:
                    return True;
    return False;

Checkoddproduct(l)  #True implies that there is a distinct pair of numbers whose product is odd.If this is not there, output is false.
```

```python
#Question 4
print("Enter the string:")  
a=input('');            #Enter the string in the console.

def vowelcounter(s):
    s=s.lower();
    count=0;
    vowels="aeiou"
    for i in s:             #The function compares the input string with a string of vowels to count the number of vowels.
        if i in vowels:
            count+=1;
    return(count)

print(vowelcounter(a))
```

```python
#Question 6 (Project)
import random


def rand(first, last, num): #Function for generating random numbers in a range.(first,last) specifies the range.(num) is number of random numbers required.
    randlist = []               #Output is in the form of a list.
  
    for j in range(num): 
        randlist.append(random.randint(first, last)) 
  
    return randlist

def fact(n):        #Function for finding factorial of a number n.
    factorial=1;    #I have used this function for finding theoretical values of probability later.
    for i in range(2,n+1):
        factorial=factorial*i
    return factorial


c=[0]*40;  #This list is used for calculating how many times in 10000 iterations atleast one same birthday occurs.
           #The list contains 40 elements for each n=5,10 and so on till 200.(Hence length 40).
for i in range(10000):  #10000 iterations for which probability in percentage is calculated.
    l=[];
    l=rand(1,365,200); #Random integers are generated using the function I have created above.
    flag=0;
    for j in range(5,201,5):     
        l1=list(l[0:j]);        #I have reused the same random list l for n= 5 to 200 by taking subsections of list l into l1.
        for k in range(len(l1)):#As I have reused lists, the execution time has been reduced significantly.
            if l1.count(l1[k])>1:
                flag=1;
                for d in range(int(j/5)-1,40):  #If a birthday is repeated then we add 1 to all j/5 -1 to 40th indices in c.
                    c[d]+=1;
                break;      #We break out of both the inner for loops after this as we already know that a birthday is repeating and the increment has been done in c for all the indexes greater than j/5-1. 
        if flag==1:         #Hence we dont need to check for more repeated birthdays in this particular list.And now after breaking out the next random integer list is generated.
            break;


expt_probab=[]   
for i in range(len(c)):          #List of experimental probabilities for all n=5 to 200 for the above experiment is created.
    expt_probab.append(c[i]/10000);
    
theor_probab=[]
f=fact(365)

for i in range(len(c)):          #List of theoretical probabilities for all n=5 to 200 is created(The formula for this can be derived and is also there on wikipedia).
    theor_probab.append(1-((f)/((365**((i+1)*5))*fact(365-((i+1)*5)))));


print("The probabilities are:") 


print("  n      experiment probability       theoretical probability")

    
for i in range(len(expt_probab)):
    print('{0:3d} :  {1:15.4f} {2:30.4f}'.format((i+1)*5,expt_probab[i],theor_probab[i])) #Probability is printed for each n.


import matplotlib.pyplot as plt

n=range(0,201,5)              #Code for plotting graph for the probabilities.
expt_probab.insert(0,0)       #Inserted n=0, probability=0 for plotting the graph.
theor_probab.insert(0,0)
plt.plot(n,expt_probab,'-')
plt.plot(n,theor_probab,'-.')
plt.xlabel('n')
plt.ylabel('proabability')
plt.legend(['experimental proabability','theoretical probability'])

print('From the graph plotted and the values shown above, it is clear that experimental and theoretical values are almost equal.')
```
